# Exercise 1 (Fill in the blanks)

```python
from matplotlib import pyplot as plt
# Some data for plotting
x = [0, 1, 2, 3, 4, 5]
y = [0, 4, 8, 12, 16, 20]
y_err = [0.0, 0.5, 1.0, 1.5, 2.0, 2.5]
x_err = .25

plt.errorbar(x, y, 
             xerr=x_err, yerr=y_err,
             color='maroon', linestyle='dashed',
             marker='o', markerfacecolor='yellow',
             ecolor='black',capsize=5,
             label='Y values with errors')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='lower right')
plt.show()
```

# Exercise 2 (A simple mathematical plot)

```python
import numpy as np
x = np.linspace(0, 50, num=100000)
y = np.exp(-x/10)*np.sin(x)

plt.style.use('default')
plt.plot(x, y, label='y', c='tab:blue', lw=1.5)
plt.xlabel('x', fontsize=15, c='darkorange') #x label fontsize & colour
plt.ylabel('y', fontsize=15, c='darkorange', rotation=0) #y label fontsize & colour
plt.grid(alpha=.25) #25% grid opacity
plt.title('$y=e^{-x/10}\\mathrm{sin(}x\\mathrm{)}$', fontsize=20) #Title
plt.ylim([-1, 1]) #Change y limit
```
